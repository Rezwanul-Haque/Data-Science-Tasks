## Importing Necessary Libraries for Data Exploration

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set() # Apply the default seaborn theme, scaling, and color palette

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

## About Dataset
This classic dataset contains the prices and other attributes of almost 54,000 diamonds. It's a great dataset for beginners learning to work with data analysis and visualization.

### Attributes

* $price =$ price in US dollars (\\$326--\\$18,823)
* $carat =$ weight of the diamond (0.2--5.01)
* $cut =$ quality of the cut (Fair, Good, Very Good, Premium, Ideal)
* $color =$ diamond colour, from J (worst) to D (best)
* $clarity =$ a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
* $x =$ length in mm (0--10.74)
* $y =$ width in mm (0--58.9)
* $z =$ depth in mm (0--31.8)
* $depth =$ total depth percentage = $z / mean(x, y) = 2 * z / (x + y) (43--79)$
* $table =$ width of top of diamond relative to widest point (43--95)

## Load the Dataset

```python
diamonds = sns.load_dataset("diamonds")  # Loaded in a Pandas DataFrame
diamonds.head()
```

## Histograms

```python
# plot
f, axes = plt.subplots(2, 2, figsize=(15, 15), sharex=False)
sns.distplot( diamonds["price"] , color="skyblue", ax=axes[0, 0])
sns.distplot( diamonds["carat"] , color="olive", ax=axes[0, 1])
sns.distplot( diamonds["depth"] , color="gold", ax=axes[1, 0])
sns.distplot( diamonds["table"] , color="teal", ax=axes[1, 1])
```

## Box Plot

```python
# plot
plt.figure(figsize=(10,10))
plot = sns.boxplot(x=diamonds['cut'], y=diamonds['price'], hue=diamonds['color'])
```

### 2.2 For categorical variables: frequency table, count plot

```python
plot = sns.countplot(diamonds['color'], hue=diamonds['cut'])
```

## 3.0 Bi-variate Analysis

### 3.1 Continuous \& Continuous: scatter plots

```python
plt.figure(figsize=(10,10))
plot = sns.scatterplot(x=diamonds['carat'], y=diamonds['price'], hue=diamonds['clarity'])
```

### 3.2 Categorical & Categorical: stacked column chart

```python
plot = (diamonds
        .groupby(['color','cut', 'clarity'])
        .size()
        .unstack()
        .plot
        .bar(stacked=True))
```

### 3.3 Categorical & Continuous: boxplots

```python
# plot
plt.figure(figsize=(10,10))
plot = sns.boxplot(x=diamonds['cut'], y=diamonds['depth'], hue=diamonds['color'])
```
